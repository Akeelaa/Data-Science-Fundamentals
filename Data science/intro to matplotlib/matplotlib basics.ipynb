```python
import matplotlib.pyplot as plt
import numpy as np

#created line graph for continous vs continous
x= np.linspace(0, 10, 100)
y= 4 + 2 * np.sin(2* x)

plt.title('Line Graph') # adding title
plt.xlabel('X axis')# assigning name to x axis
plt.ylabel('Y Axis') # adding a name to y axis
plt.plot(x,y) # this prints it out

```

```python
#Scatter Graph

x = 4 + np.random.normal(0,2,24)
y = 4 + np.random.normal(0,2, len(x))

plt.scatter(x,y)

```
