```python
# Import pandas
import pandas as pd
```

```python
# Create a DataFrame with balance.txt

df = pd.read_csv('balance.txt', delim_whitespace=True)
```

```python
df.iloc[5:10]
```

```python
# Compare the average income based on ethnicity.
print("The average income for Caucasian is", df[df.Ethnicity == "Caucasian"].loc[:,"Income"].mean())
print("The average income for African American is", df[df.Ethnicity == "African American"].loc[:,"Income"].mean())
print("The average income for  Asian is", df[df.Ethnicity == "Asian"].loc[:,"Income"].mean())

```

```python
# On average, do married or single people have a higher balance? Single people

print(df.groupby('Married')['Balance'].mean(numeric_only=True))
```

```python
# What is the highest income in our dataset?
highest_income = df['Income'].max()
print("max income is:", highest_income)

```

```python
# What is the lowest income in our dataset?

min_income = df['Income'].min()

print("max income is:", min_income)
```

```python
# How many cards do we have recorded in our dataset? (Hint: use `sum()`)

df['Cards'].sum()
```

```python
# How many females do we have information for vs how many males? (Hint: use `count()

df['Gender'].value_counts()
```
