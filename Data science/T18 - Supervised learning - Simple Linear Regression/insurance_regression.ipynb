```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
%matplotlib inline


df = pd.read_csv('insurance.csv')
```

```python
# 1. Scatter plot of age vs insurance charges

x = df.iloc[:, :1].values  #   Im getting the age data puting in x variable
y = df.iloc[:, -1].values  # Im getting the insrance charges data
plt.scatter(x, y, color='b')  # Giving the Scatter plot with blue color
plt.xlabel('Age')  # Label for x-axis
plt.ylabel('Insurance charges')  # Label for y-axis
plt.show()  # This will display the scatter plot

# 2. Creating a Linear Regression model

salary_model = LinearRegression()  # Creating an instance of LinearRegression
salary_model.fit(x, y)  # Fitting the model using age as input and insurance charges as output

# 3. Plotting the regression line on the scatter plot

y_pred = salary_model.predict(x)  # Predicting insurance charges based on the age using the trained model

plt.scatter(x, y, color='b')  # Scatter plot with blue color
plt.plot(x, y_pred, color='r')  # Plotting the regression line with red color
plt.title('How age affects insurance charges')  # Title of the plot
plt.ylabel('Insurance charges')  # Label for y-axis
plt.xlabel('Age')  # Label for x-axis
plt.show()  # This will display the scatter plot with the regression line

```
